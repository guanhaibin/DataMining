```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.preprocessing import OnehotTransactions
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
meals=[['M1','M2','M5'],
      ['M2','M4'],
      ['M2','M3'],
      ['M1','M2','M4'],
      ['M1','M3'],
       ['M2','M3'],
      ['M1','M3'],
      ['M1','M2','M3','M5'],
      ['M1','M2','M3'],
      ['M1','M3','M4','M5']]
print(meals)

##encode data into OnehotTransaction dataset
oht=OnehotTransactions()
oht_array=oht.fit(meals).transform(meals)
df=pd.DataFrame(oht_array,columns=oht.columns_)
df
```

###### a. Identify all the frequent k-itemsets 

```python
###generate the frequent itemsets by given min_support=0.2
frequent_itemsets=apriori(df,min_support=0.2,use_colnames=True)
frequent_itemsets
```

```python
###find out the candidates that are pruned due to a violation of the Apriori property
###generate all candidates before prune
before_prune=apriori(df,min_support=0,use_colnames=True)
before_prune['length']=before_prune['itemsets'].apply(lambda x:len(x))
before_prune
```

Notice, for length 2 candidates: 

(M3, M4) 

(M5, M4) 

would be pruned since they didn't pass the min_support threshold, 

thus, for length 3 candidates:

(M3,M4,M1),

(M5,M4,M1),

(M2,M3,M4),

(M5,M2,M4),

(M5,M3,M4), 

they were not generated since there were no (M3,M4) and (M5,M4).

Further, for length 4 candidates, 

(M2,M4,M3,M1), 

(M5,M2,M3,M1), 

(M5,M2,M4,M1),

(M5,M3,M4,M1),

(M5,M2,M3,M4),

and length 5 candidate:

(M5,M4,M2,M3,M1)

were not be generated neither since there were missing those length 3 candidates.

##### b. find all strong association rules of the form X^Y-Y

```python
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=0.7)
rules['antecedent_len']=rules['antecedents'].apply(lambda x:len(x))
print("The assoication rules of the form X^Y -> Y that confidences are higher than the 0.7 are:")
rules[rules['antecedent_len']>=2][['antecedents','consequents','confidence']]
```
